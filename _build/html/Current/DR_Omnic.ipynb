# Omnic Processing

When the samples have not been saved (and exported) correctly, the data obtained can be corrupted and needs to be repoened with OMNIC to be resaved and exported as csv. However OMNIC is a commercial, software (ThermoFisher), and an alternative needs to be found to be compliant with `Open Research` 

- A good library to use is **Spectrochempy** [Doc](https://www.spectrochempy.fr/stable/gettingstarted/install/install_win.html)

```python
import spectrochempy as scp
```

## Import and Read data

```python
date = "2020_09_28"
```

```python
Scan = scp.read_omnic("..\..\Data\DATA_RAW\{}\ASW_{}_0044.spa".format(date,date))
```
