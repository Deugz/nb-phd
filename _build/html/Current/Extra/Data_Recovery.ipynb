## Loop through df Column

```python
nscan = len(list(df_2021_10_18.columns.values.tolist()))

print(nscan)

file_number = 1

for column in df_2021_10_18:
    
    print(file_number)
    #print(df_2020_09_28[column])
    
    df = df_2021_10_18[column]
    
    df1 = pd.DataFrame(df)
    
    df_W = df1.join(Wavenumber)
    
    df_W.insert(0, 'Wavenumber', df_W.pop('Wavenumber'))
    
    #print(df_W)
    
    if file_number in range(1, 10):
    
        df_W.to_csv("..\..\Data\DATA_PROCESSING\{}\Data\Recovery\C2H6_ASW_{}_000{}_smooth.csv".format(date,date,file_number), index=False, header=False)
        
    elif file_number in range(10, 100):
    
        df_W.to_csv("..\..\Data\DATA_PROCESSING\{}\Data\Recovery\C2H6_ASW_{}_00{}_smooth.csv".format(date,date,file_number), index=False, header=False)
        
    else:
        
        df_W.to_csv("..\..\Data\DATA_PROCESSING\{}\Data\Recovery\C2H6_ASW_{}_0{}_smooth.csv".format(date,date,file_number), index=False, header=False)
    
    file_number +=1

```
