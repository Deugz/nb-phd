## Loading a full dataset 

Create a list of files using glob function and implement them into list_files.

```python
list_files = []

file_path = "..\..\Data\DATA_RAW\{}\ASW_{}_*.spa".format(date,date)

for file in glob.glob(file_path):
    
    list_files.append(file) 
```

```python
#list_files
```

```python
Data = scp.read_omnic(list_files, directory="..\..\Data\DATA_RAW\{}".format(date))
```
